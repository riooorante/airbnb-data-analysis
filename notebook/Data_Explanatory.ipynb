```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data_DF = pd.read_csv("/content/sample_data/airbnb.csv")
Data_DF
```

```python
Data_DF.info()
```

```python
Data_DF.describe()
```

```python
Data_DF.dropna(inplace = True)
Data_DF.info()
```

```python
sns.heatmap(Data_DF.corr(), annot=True, cmap="YlGnBu")
```

```python
Data_DF.columns
```

```python
Data_DF['ds'] = pd.to_datetime(Data_DF['ds'])
Data_DF['ts_min'] = pd.to_datetime(Data_DF['ts_min'])
Data_DF['ts_max'] = pd.to_datetime(Data_DF['ts_max'])
```

```python
Data_DF
```

```python
Data_DF['year'] = Data_DF['ds'].dt.year
Data_DF['month'] = Data_DF['ds'].dt.month
Data_DF['day'] = Data_DF['ds'].dt.day
Data_DF['hour'] = Data_DF['ts_min'].dt.hour
Data_DF
```

```python
pengunjung_perhari = Data_DF.groupby(Data_DF['ds'].dt.date)['id_visitor'].nunique()

pengunjung_perhari.plot(kind='line', figsize=(10, 6))
plt.title('Jumlah Pengunjung per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```

```python
pengunjung_berdasarkan_device = Data_DF.groupby(Data_DF["dim_device_app_combo"])['id_visitor'].nunique()

pengunjung_berdasarkan_device.plot(kind='line', figsize=(30, 6))
plt.title('Device yang digunakan')
plt.xlabel('Device')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```

```python
pengunjung_berdasarkan_device.sort_values()
```

```python
pengunjung1 = Data_DF.groupby(Data_DF["sent_booking_request"])['id_visitor'].nunique()

pengunjung1.plot(kind='line', figsize=(10, 6))
plt.title('Pengunjung yang Booking 2')
plt.xlabel('Booking (1.0)')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```

```python
pengunjung2 = Data_DF.groupby(Data_DF["next_sent_booking_request"])['id_visitor'].nunique()

pengunjung2.plot(kind='line', figsize=(10, 6))
plt.title('Pengunjung yang Booking 2')
plt.xlabel('Booking (1.0)')
plt.ylabel('Jumlah Pengunjung')
plt.show()
```
